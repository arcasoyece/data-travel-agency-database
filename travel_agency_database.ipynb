### Bir fonksiyon oluşturun

`city` ve `key` olmak üzere iki argüman alan ve şehir için karşılık gelen değeri döndüren `get_key_value_from_city` fonksiyonunu oluşturun.

__Örnek__: `get_key_value_from_city(city=paris_city, key="total_population")` 2102650 döndürecektir.

<u>Not:</u> Eğer bir anahtar (key) yoksa, fonksiyon `"_<key>_ not defined"` döndürmelidir.
    




```python
def get_key_value_from_city(city, key):
    # Eğer anahtar (key) sözlüğün içinde varsa, değeri döndür
    if key in city:
        return city[key]
    # Eğer anahtar yoksa, "_<key>_ not defined" hata mesajını döndür
    else:
        return f"_{key}_ not defined"

# Örnek 
print(get_key_value_from_city(city=paris_city, key="total_population"))

print(get_key_value_from_city(city=paris_city, key="landmark"))
```

## Görev 2 - Tüm şehirler için bilgi alma



Seyahat acentesi, konaklama sağladığı birden fazla şehir hakkında bilgi topladı. Bu şehirlerin tüm bilgileri `cities` adlı bir Python listesinde saklanır.

Aşağıdaki hücreyi çalıştırarak verileri yükleyin

```python
cities = [
    {
        "name": "Paris",
        "total_population": 2102650,
        "continent": "Europe",
        "location": [48.8566, 2.3522],
        "area": {
            "prefecture_level": 105.4
        }
    },
    {
        "name": "Xi’an",
        "total_population": 12328000,
        "continent": "Asia",
        "location": [34.2667, 108.9000],
        "area": {
            "prefecture_level": 10762
        }
    },
    {
        "name": "Chengdu",
        "total_population": 14645000,
        "continent": "Asia",
        "location": [30.6600, 104.0633],
        "area": {
            "prefecture_level": 14378
        }
    },
    {
        "name": "Los Angeles",
        "total_population": 12121244,
        "continent": "North America",
        "location": [34.1141, -118.4068],
        "area": {
            "prefecture_level": 1299
        }
    },
    {
        "name": "Brooklyn",
        "total_population": 2736074,
        "continent": "North America",
        "location": [40.6501, -73.9496],
        "area": {
            "prefecture_level": 180
        }
    }
]
```

Onların kaç şehri var?

```python
sehir_sayisi = len(cities)
print(sehir_sayisi)
```

### Değerleri okuma


Önceki `get_key_value_from_city` fonksiyonunuzu kullanarak iki argüman alan yeni bir fonksiyon oluşturun: `city_name` ve `key`.

Eğer şehir listede yoksa, fonksiyon `"*city* doesn't exist or is not covered"` döndürmelidir.

Fonksiyonunuzu çeşitli örneklerle test edin.

```python
def get_key_value_from_city_name(city_name, key):
    # 1. Cities listesi içinde her bir şehir sözlüğünü kontrol et
    for city in cities:
        # 2. Eğer o anki şehrin ismi aranan city_name ise
        if city["name"] == city_name:
            # Önceki fonksiyonumuzu kullanarak veriyi döndür
            return get_key_value_from_city(city, key)
    
    # 3. Eğer döngü biter ve şehir bulunamazsa hata mesajı döndür
    return f"_{city_name}_ doesn't exist or is not covered"

# Örnek Testler:
print(get_key_value_from_city_name("Paris", "total_population"))
# Sonuç: 2102650

print(get_key_value_from_city_name("Xi’an", "continent"))
# Sonuç: Asia

print(get_key_value_from_city_name("Istanbul", "total_population"))
# Sonuç: _Istanbul_ doesn't exist or is not covered
```

### Değer güncelleme


Bazen şehir bilgileri yanlış olabilir veya güncellemeye ihtiyaç duyabilir.

`modify_city_key_value` fonksiyonunu oluşturun; bu fonksiyon mevcut bir anahtarın değerini yeni bir değerle değiştirir. \
Fonksiyon üç argüman alır: `city_name`, `key` ve anahtara atanacak yeni `value`.

Fonksiyon aşağıdaki durumlardan birini belirten bir metin mesajı üretecektir:

- Eğer şehir yoksa veya kapsanmıyorsa: `"*city* doesn't exist or is not covered."`
- Eğer verilen anahtar yoksa: `"*key* doesn't exist."`
- Eğer anahtar başarıyla güncellendiyse: `"*key* has been updated for *city*."`

```python
def modify_city_key_value(city_name, key, value):
    # 1. Şehri bulmak için listeyi tara
    for city in cities:
        if city["name"] == city_name:
            # 2. Şehir bulundu, şimdi anahtar (key) var mı kontrol et
            if key in city:
                # 3. Anahtar varsa değeri güncelle
                city[key] = value
                return f"{key} has been updated for {city_name}"
            else:
                # Anahtar şehirde bulunamadı
                return f"{key} doesn't exist"
    
    # 4. Döngü bitti ve şehir bulunamadıysa
    return f"{city_name} doesn't exist or is not covered"

# Fonksiyon testi:
print(modify_city_key_value("Paris", "total_population", 2200000))
# Sonuç: total_population has been updated for Paris

print(modify_city_key_value("Istanbul", "total_population", 15000000))
# Sonuç: Istanbul doesn't exist or is not covered

print(modify_city_key_value("Xi’an", "landmark", "Terracotta Army"))
# Sonuç: landmark doesn't exist
```

### Arama motorunu tasarlayın


`get_cities_nearby` adlı bir fonksiyon oluşturun; bu fonksiyon iki argüman alır:
- `customer_location`: [lat, long] formatında koordinatlardan oluşan bir liste


- `radius`: kilometre cinsinden eşik değeri

Ve verilen `customer_location` yakınında bulunan `şehirlerin bir listesini` döndürür.

Eğer hiçbir şehir bulunamazsa, fonksiyon boş bir liste döndürmelidir.

Test etmek için Çin'den Guangyan şehrini ([32.4355, 105.8436]) ve 500 km'lik bir yarıçapı kullanabilirsiniz.

Xi'an ve Chengdu'nun çok uzak olmadığını görmelisiniz...

```python
def get_cities_nearby(customer_location, radius):
    nearby_cities = []  # Yakın şehirleri toplamak için boş bir liste
    
    # 1. Listedeki tüm şehirleri tek tek dönüyoruz
    for city in cities:
        city_location = city["location"]
        
        # 2. Şehir ile müşteri arasındaki mesafeyi hesaplıyoruz
        dist = distance.distance(customer_location, city_location).km
        
        # 3. Eğer mesafe verilen yarıçaptan (radius) küçük veya eşitse
        if dist <= radius:
            nearby_cities.append(city["name"])
            
    # 4. Yakındaki şehirlerin isimlerinden oluşan listeyi döndür
    return nearby_cities

# --- Test ---
guangyan_location = [32.4355, 105.8436]
nearby = get_cities_nearby(guangyan_location, 500)

print(nearby)
```
